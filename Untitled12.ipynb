```python
import speech_recognition as sr
reco=sr.Recognizer()
with sr.AudioFile('/content/audio.wav') as source:
  audio_data=reco.record(source)
  text=reco.recognize_google(audio_data)
  print('transcript:',reco.recognize_google(audio_data))
```
